```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
```

**QUESTÃO 1** - Mola amortecedor

```python
# Parâmetros do Sistema
m1, m2 = 20, 20
k1, k2 = 1000, 2000
b1, b2 = 1, 8

# Definindo as equações diferenciais
def system(y, t):
    x1, v1, x2, v2 = y

    dx1dt = v1
    dv1dt = (-b1*v1 - k1*x1 + k2*(x2 - x1))/m1
    dx2dt = v2
    dv2dt = (-b2*v2 - k2*(x2 - x1))/m2

    return [dx1dt, dv1dt, dx2dt, dv2dt]

# Condições iniciais
aux = [1, 0, 2, 1]

# Tempo de simulação
t = np.linspace(0, 50, 10000)

# Resolvendo as equações diferenciais
sol = odeint(system, aux, t)

# Extraindo as soluções
x1, v1, x2, v2 = sol.T

# Subplots para os gráficos de posição e velocidade
plt.figure(figsize=(10, 4))
plt.subplot(2, 2, 1)
plt.plot(t, x1, 'r', label='x1(t)')
plt.title('Posição de m1 ao longo do tempo')
plt.xlabel('Tempo')
plt.ylabel('Posição')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(t, v1, 'b', label='v1(t)')
plt.title('Velocidade de m1 ao longo do tempo')
plt.xlabel('Tempo')
plt.ylabel('Velocidade')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(t, x2, 'g', label='x2(t)')
plt.title('Posição de m2 ao longo do tempo')
plt.xlabel('Tempo')
plt.ylabel('Posição')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(t, v2, 'm', label='v2(t)')
plt.title('Velocidade de m2 ao longo do tempo')
plt.xlabel('Tempo')
plt.ylabel('Velocidade')
plt.legend()

plt.tight_layout()

# Subplots para os diagramas de fases
plt.figure(figsize=(8, 4))

# Diagrama de Fases 1
plt.subplot(1, 2, 1)
plt.plot(x1, v1, 'r', label='m1')
plt.title('Diagrama de Fases 1')
plt.xlabel('Posição')
plt.ylabel('Velocidade')
plt.legend()

# Diagrama de Fases 2
plt.subplot(1, 2, 2)
plt.plot(x2, v2, 'b', label='m2')
plt.title('Diagrama de Fases 2')
plt.xlabel('Posição')
plt.ylabel('Velocidade')
plt.legend()

plt.tight_layout()
plt.show()

# Diagrama de Fases 1 em 3D
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot(t, x1,v1, 'r', label='m1')
ax1.set_title('Diagrama de Fases 1')
ax1.set_xlabel('Posição')
ax1.set_ylabel('Velocidade')
ax1.set_zlabel('Tempo')
ax1.legend()

# Diagrama de Fases 2 em 3D
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot(t, x2,v2, 'b', label='m2')
ax2.set_title('Diagrama de Fases 2')
ax2.set_xlabel('Posição')
ax2.set_ylabel('Velocidade')
ax2.set_zlabel('Tempo')
ax2.legend()

plt.tight_layout()
plt.show()
```

**QUESTÃO 2** Predador vs Presa

```python
# Função do Sistema de equações diferenciais
def system(y, t, a, b, c, d, e, h):
    x1, y1 = y
    dxdt = x1 * (a - c * x1 - d * y1)
    dzdt = -y1 * (b - e * x1) - h
    return [dxdt, dzdt]

# Encontrar pontos críticos
def p_critico(a, b, c, d, e, h):
    x_critico = [0, (b/e)]
    y_critico = [0, (a/c)]
    return x_critico, y_critico

# Parâmetros dados
a = 1
b = 0.25
c = 0.01
d = 0.02
e = 0.02
h = 0

presa = [1, 5, 10, 28, 7, 9]

# Criar grade de valores para x e y
x = np.linspace(0, 70, 25)
y = np.linspace(0, 70, 25)
X, Y = np.meshgrid(x, y)

# Calcular as taxas de mudança (x' e y') para cada ponto na grade
dxdt, dydt = system([X, Y], 0, a, b, c, d, e, h)

# Encontrar os pontos críticos
x_critico, y_critico = p_critico(a, b, c, d, e, h)
print("Pontos Críticos do Sistema:")
for i in range(len(x_critico)):
    print(f"Ponto Crítico {i + 1}: X={round(x_critico[i], 2)}, Y={round(y_critico[i], 2)}")

# Gráfico do Sistema 1
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, dxdt, dydt, color='black', angles='xy', scale_units='xy')
plt.xlabel('Presas (x)')
plt.ylabel('Predador (y)')
plt.title('Retrato de Fase do Sistema Lotka-Volterra')
plt.xlim([0, 70])
plt.ylim([0, 70])
plt.grid(True)

# Simulação da dinâmica do Sistema para diferentes condições iniciais
for pre_sa  in presa:
    y0 = [15, pre_sa ]
    linha_pontos = np.linspace(0, 50, 200)
    ys = odeint(system, y0, linha_pontos, args=(a, b, c, d, e, h))
    plt.plot(ys[:, 0], ys[:, 1], label=f'y0={pre_sa }')
plt.legend()
plt.show()


# Gráfico do Sistema Caça permitida
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, dxdt, dydt, color='black', angles='xy', scale_units='xy')
plt.xlabel('Presas (x)')
plt.ylabel('Predador (y)')
plt.title('Caça permitida')
plt.xlim([0, 70])
plt.ylim([0, 70])
plt.grid(True)

# Simulação da dinâmica do Sistema para diferentes condições iniciais
for pre_sa  in presa:
    y0 = [pre_sa *2, pre_sa ]
    linha_pontos = np.linspace(0, 50, 200)
    ys = odeint(system, y0, linha_pontos, args=(a, b, c, d, e, h))
    plt.plot(ys[:, 0], ys[:, 1], label=f'y0={pre_sa}')

plt.legend()
plt.show()
```

**QUESTÃO 3** Pendulo

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parâmetros do Sistema
L = 2
L0 = 1
k = 40
m = 1
g = 9.81
theta = (3 * np.pi) / 4

# Condições iniciais
x1_0 = 0  # Posição inicial
x2_0 = 0  # Velocidade inicial

# Sistema de equações diferenciais
def system(y, t):
    x1, x2 = y
    dxdt = (-1 / L) * (g * np.sin(theta) + 2 * x1 * x2)
    dydt = (1 / m) * (m * L * x1 ** 2 - k * (L - L0) + m * g * np.cos(theta))

    return [dxdt, dydt]

# Tempo
t = np.linspace(0, 9, 1000)

# Resolvendo as equações diferenciais
sol = odeint(system, [x1_0, x2_0], t)

# Extraindo as soluções
x1 = sol.T[0]
x2 = sol.T[1]

# Gráficos
plt.figure(figsize=(10, 10))

# Gráfico da posição em função do tempo
plt.subplot(3, 1, 1)
plt.plot(t, x1, 'b', label='Posição (x1)')
plt.xlabel('Tempo')
plt.ylabel('Posição')
plt.legend()

# Gráfico da velocidade em função do tempo
plt.subplot(3, 1, 2)
plt.plot(t, x2, 'r', label='Velocidade (x2)')
plt.xlabel('Tempo')
plt.ylabel('Velocidade')
plt.legend()

# Diagrama de fase
plt.subplot(3, 1, 3)
plt.plot(x1, x2, 'g', label='Diagrama de Fase')
plt.xlabel('Posição')
plt.ylabel('Velocidade')
plt.legend()

# Adicionando setas ao diagrama de fase
x1_range = np.linspace(min(x1), max(x1), 20)
x2_range = np.linspace(min(x2), max(x2), 20)
X1, X2 = np.meshgrid(x1_range, x2_range)
dxdt, dx2dt = system([X1, X2], 0)
norm = np.sqrt(dxdt**2 + dx2dt**2)
dxdt /= norm
dx2dt /= norm
plt.quiver(X1, X2, dxdt, dx2dt, color='r', scale=75, width=0.001)

plt.tight_layout()
plt.show()
```
